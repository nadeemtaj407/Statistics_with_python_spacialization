Lets say we would like to splice our data frame and select only specific portions of our data. 

There are three different ways of doing so.

.loc()
.iloc()
.ix()
We will cover the .loc() and .iloc() splicing functions.

.loc()
.loc() takes two single/list/range operator separated by ','. The first one indicates the row and the second one indicates columns.

##.loc()

```python
# Return all observations of batter
df.loc[:, "batter"]
```

```python
#select all rows of multiples columns ["innings", "overs", "ballnumber"] 
df.loc[:, ["innings", "overs", "ballnumber"]]
```

```python
#select 10 rows of multiples columns ["innings", "overs", "ballnumber"] 
df.loc[:9, ["innings", "overs", "ballnumber"]]
```

```python
# Select range of rows for all columns
df.loc[5:10]
```

```python
# Firts 5 rows of any specific columns
df.loc[:5, "innings"]
```

##.iloc()

.iloc() is integer based slicing, whereas .loc() used labels/column names. 

```python
df.iloc[:4]
```

```python
#specific rows and columns
df.iloc[1:10, :3 ]
```

```python
#top 5 scorer in ipl 2022
Score=df.groupby(['batter'])[['batsman_run']].sum()
Score.reset_index(inplace=True)
plt.figure(figsize=(20,6))
plt.title('Top 3 scorer in IPL 2022',size=15,fontweight='bold')
plt.xticks(rotation = 90)
plt.grid(visible=True)
ax=sns.barplot(data=Score.sort_values('batsman_run',ascending=False).head(5),
            x='batter',y='batsman_run',palette='viridis')
for i in ax.containers:
    ax.bar_label(i,)
```
